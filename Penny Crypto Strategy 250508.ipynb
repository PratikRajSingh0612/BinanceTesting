```python
test = CreateVars_AllPairs(top_volume_pairs, Pairs_CandlesData, client, BackTime)

test
```

```python
# Calculate scores
test['volatility_score'] = (test['ATR'] / test['Close']) * 100  # Normalized ATR
test['momentum_score'] = np.where(
    test['RSI'] > 70, 1, np.where(test['RSI'] < 30, -1, 0)
)
test['volume_score'] = test['Quote Volume'].rank(pct=True)

# Combine scores (adjust weights here)
test['total_score'] = (
    0.4 * test['volatility_score'] +
    0.3 * test['momentum_score'] +
    0.3 * test['volume_score']
)

# Sort and format
test1 = test.sort_values('total_score', ascending=False).head(10)
test1 = test1[['Symbol', 'total_score', 'volatility_score', 'momentum_score', 'volume_score']]
test1
```
